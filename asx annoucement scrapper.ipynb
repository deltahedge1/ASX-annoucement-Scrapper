```python
import pandas as pd
import matplotlib as plt
import requests
from bs4 import BeautifulSoup
import re
```

```python
def clean_text(text):
    temp=[]
    new=text.split()
    for i in new:
        if i != "":
            temp.append(i)
    
    return(temp)
```

```python
r = requests.get(r'http://m.asx.com.au/m/announcements.xhtml')
sauce = r.text
soup = BeautifulSoup(sauce, "lxml")
```

```python
print(soup.prettify())
```

```python
url_path = r'http://www.asx.com.au/'
p = re.compile(r"(?:encodeURIComponent\('(.+?)'\))")
td = soup.find_all("td")

list1 =[]
for j in td:
    
    text = clean_text(j.text)
    list1.append(text)
    tx = str(j.find("a"))
    
    for i in p.findall(tx):
        href = clean_text(url_path+i)
        list1.append(href)

#list1        
#http://www.asx.com.au//asxpdf/20171020/pdf/43nfkmx30cs5hq.pdf
```

```python
list2=[]
for i in list1:
    if len(i)>1:
        list2.append(" ".join(i))
    else:
        list2.append(i[0])

list3 = []
k = 0
for j in list2[::4]:
    
    temp = []
    
    temp.append(list2[k])
    temp.append(list2[k+1])
    temp.append(list2[k+2])
    temp.append(list2[k+3])
    
    k+=4
    list3.append(temp)
    
list3
```
